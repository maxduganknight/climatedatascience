```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../../data/UNSEEN/heatwave/food_prices/corn_prices_historical.csv')
df.head()
```

```python
# Filter the dataframe for the years 2001-2011
filtered_df = df[(df['date'] >= '2002-05-01') & (df['date'] <= '2012-05-31')]

# Calculate the mean of the 'value' column
mean_value = filtered_df['value'].mean()

mean_value
```

```python
# Filter the dataframe for the date range June - December 2012
filtered_df_2012 = df[(df['date'] >= '2012-06-01') & (df['date'] <= '2012-12-31')]

# Calculate the mean of the 'value' column
mean_value_2012 = filtered_df_2012['value'].mean()

mean_value_2012
```

```python
(mean_value_2012-mean_value) / mean_value
```
