```python
import pandas as pd
import numpy as np

import glob
import sys
import os
from census import Census
from us import states


# creds
sys.path.append('/Users/max/Deep_Sky')
from creds import CENSUS_API_KEY
```

```python
c = Census(CENSUS_API_KEY)
```

```python
fields = ['NAME', 'B01003_001E']  
state_fips_codes = [state.fips for state in states.STATES]

data = []
for state_fips in state_fips_codes:
    counties = c.acs5.state_county(fields, state_fips, '*')
    data.extend(counties)

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Split the 'NAME' column into 'County Name' and 'State Name'
df[['county_name', 'state_name']] = df['NAME'].str.split(', ', expand=True)

# Rename columns and reorder
df.rename(columns={
            'B01003_001E': 'population',
            'state': 'state_fips',
            'county': 'county_fips'
        }, inplace=True)

df = df[['state_name', 'state_fips', 'county_name', 'county_fips', 'population']]

df.head()
```

```python
df.to_csv('../data/us_census/county_populations.csv', index=False)
```
