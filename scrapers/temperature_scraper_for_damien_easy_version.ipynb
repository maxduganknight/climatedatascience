```python
import requests
import pandas as pd
import numpy as np
import io
import plotly.express as px
```

```python
temperature_file_url = 'https://sites.ecmwf.int/data/c3sci/bulletin/202407/temperature/timeseries_era5_monthly_2t_global_allmonths_up_to_july_2024.csv'
```

```python
urlData = requests.get(temperature_file_url).content
df = pd.read_csv(io.StringIO(urlData.decode('utf-8')), skiprows=11)
df
```

```python
df_clean = df.copy()
df_clean['month'] = pd.to_datetime(df_clean['month'])
df_clean['year'] = df_clean['month'].dt.year
df_clean['month'] = df_clean['month'].dt.month

# Pivot the DataFrame
pivot_df = df_clean.pivot(index='month', columns='year', values='ano_pi')

# Rename columns to month names if desired

year_strings = [str(year) for year in range(1940, 2025)]

pivot_df.columns = year_strings
pivot_df.index = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
pivot_df
#pivot_df.to_csv('../data/global_temperatures/monthly_air_temperature_anomaly.csv')
```
